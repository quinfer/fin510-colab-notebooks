# Usage Notes

-   The readings above are deliberately selective and rigorous; they
    anchor each theme for both modules.  
-   For FIN510, use them as guided readings (focus on concepts and
    evidence).  
-   For FIN720, expect critical appraisal and application.  
-   Coding work should link to methods in the papers (e.g., ML
    baselines, text sentiment, event study design).

Ausubel, Lawrence M. 1991. “The Failure of Competition in the Credit
Card Market.” *American Economic Review*.
<https://doi.org/10.2307/2006788>.

B"ohme, Rainer, Nicolas Christin, Benjamin Edelman, and Tyler Moore.
2015. “Bitcoin: Economics, Technology, and Governance.” *Journal of
Economic Perspectives*. <https://doi.org/10.1257/jep.29.2.213>.

<https://doi.org/10.1016/j.jfineco.2018.03.011>.

<https://doi.org/10.1093/rfs/hhad021>.
